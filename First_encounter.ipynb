```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import ast
```

```python
#Read data from the gdrive
# Path to your text file
file_01 = '/content/drive/My Drive/First_encounter/x_train.txt'
file_02 = '/content/drive/My Drive/First_encounter/y_train.txt'
file_03 = '/content/drive/My Drive/First_encounter/x_test.txt'
file_04 = '/content/drive/My Drive/First_encounter/y_test.txt'
#file_05 = '/content/drive/My Drive/First_encounter/train data correct.txt'

# Read the text file
with open(file_01, 'r') as file:
    x_train = file.read()
with open(file_02, 'r') as file:
    y_train = file.read()
with open(file_03, 'r') as file:
    x_test = file.read()
with open(file_04, 'r') as file:
    y_test = file.read()

x_train = ast.literal_eval(x_train)
x_train = np.array(x_train, dtype=np.float32)

y_train = ast.literal_eval(y_train)
y_train = np.array(y_train, dtype=np.float32)

x_test = ast.literal_eval(x_test)
x_test = np.array(x_test, dtype=np.float32)

y_test = ast.literal_eval(y_test)
y_test = np.array(y_test, dtype=np.float32)

print(len(x_train[1]))
```

```python
import tensorflow as tf
import numpy as np

# Define the data
x_train_data = x_train #np.array([[1, 2, 3], [2, 3, 2], [2, 2, 1], [5, 1, 1], [3, 1, 1], [4, 4, 5]], dtype=np.float32)
x_test_data = x_test #np.array([[5, 1, 2], [4, 1, 4], [2, 1, 2]], dtype=np.float32)
y_train_data = y_train #np.array([1, 1, 1, 0, 0, 0], dtype=np.float32)
y_test_data = y_test #np.array([1, 1, 0], dtype=np.float32)

# Define the model
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(900,)),   # Input layer with 3 features
    #tf.keras.layers.Dense(64, activation='relu'),   # Hidden layer with 64 units and ReLU activation
    tf.keras.layers.Dense(1, activation='sigmoid')  # Output layer with sigmoid activation
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(x_train_data, y_train_data, epochs=100, batch_size=2, verbose=1)

# Evaluate the model on test data
loss, accuracy = model.evaluate(x_test_data, y_test_data)
print(f"Test loss: {loss:.4f}, Test accuracy: {accuracy:.4f}")

# Make predictions
predictions = model.predict(x_test_data)
rounded_predictions = np.round(predictions)

print("Predictions:")
for i in range(len(x_test_data)):
    print(f"Input: {x_test_data[i]}, Predicted: {rounded_predictions[i][0]}, Actual: {y_test_data[i]}")
```
